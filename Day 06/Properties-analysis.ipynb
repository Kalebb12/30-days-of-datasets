```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
properties_df = pd.read_csv('data/properties.csv')

# Display the first few rows of the dataset and its summary
properties_head = properties_df.head()
properties_summary = properties_df.describe(include='all')

properties_head, properties_summary
```

```python
import seaborn as sns
colors = sns.color_palette('viridis', 10)
```

```python
# Adjusting the color scheme for all three charts based on the 'viridis' palette

# Initialize the figure
plt.figure(figsize=(15, 10))

# Setting up the color palette
colors = sns.color_palette('viridis', 10)

# Plotting histograms with adjusted color scheme for Size_sqft and SalePrice
plt.subplot(2, 2, 1)
sns.histplot(properties_df['Size_sqft'], kde=True, color=colors[4], bins=10)
plt.title('Distribution of Property Size (sqft)')
plt.xlabel('Size (sqft)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(properties_df['SalePrice'], kde=True, color=colors[7], bins=10)
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')

# Plotting a bar chart for Location with the same color scheme
plt.subplot(2, 2, 3)
sns.countplot(data=properties_df, x='Location', palette='viridis')
plt.title('Distribution of Properties by Location')
plt.xlabel('Location')
plt.ylabel('Number of Properties')

plt.tight_layout()
plt.show()
```

### 1. Distribution of Property Size (sqft)
- The histogram shows how property sizes are distributed.
- There seems to be a concentration of properties around smaller sizes (1,0001,500 sqft), suggesting smaller properties are more common.
- The distribution is somewhat bimodal, indicating two groups  smaller homes and larger homes  possibly reflecting different property types (e.g., apartments vs. detached houses).

### 2. Distribution of Sale Price
- The sale prices range roughly from $200,000 to 600,000.
- The distribution is fairly even, though there might be a slight increase toward higher prices, suggesting a good number of premium properties.
- The smooth density curve indicates no extreme skewness, meaning prices are fairly well spread out.

### 3. Distribution of Properties by Location
- Suburb areas have the highest number of properties, followed by Downtown, and Countryside has the least.
- This could mean suburban properties dominate the dataset, possibly due to greater availability or development focus there.

### Overall Insights
- The dataset likely represents a balanced mix of property sizes and prices, with suburban dominance.
- Theres a potential correlation between location and size/price  suburban properties may be more numerous and moderately priced, while countryside ones could be fewer but larger.

### Relationship Analysis

```python
# Proceeding with Relationship Analysis

# Initialize the figure
plt.figure(figsize=(15, 10))

# Scatter plot between Size_sqft and SalePrice
plt.subplot(2, 2, 1)
sns.scatterplot(data=properties_df, x='Size_sqft', y='SalePrice', hue='Location', palette='viridis', s=80)
plt.title('Relationship between Property Size and Sale Price')
plt.xlabel('Size (sqft)')
plt.ylabel('Sale Price')

# Box plots to compare SalePrice across different Location categories
plt.subplot(2, 2, 2)
sns.boxplot(data=properties_df, x='Location', y='SalePrice', palette='viridis')
plt.title('Sale Price Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Sale Price')

plt.tight_layout()
plt.show()
```

## Relationship and Price Distribution Analysis

### 1. Relationship between Property Size and Sale Price
The scatter plot compares **property size (sqft)** and **sale price**, with colors representing **location**.

- Theres **no strong linear correlation** between size and price  the data points are widely scattered.  
- This suggests that **factors beyond size**, such as **location, property type, or amenities**, have a significant influence on price.  
- All three locations (Suburb, Downtown, Countryside) cover a wide range of sizes and prices, indicating that price variation is not solely dependent on property size.

---

### 2. Sale Price Distribution by Location
The boxplot shows how **sale prices vary across locations**:

- **Suburb:**  
  Has a **high median sale price** and a wide range, suggesting a mix of affordable and premium properties.  

- **Downtown:**  
  Exhibits **lower median prices** and a **narrower range**, indicating more consistent pricing  likely due to uniform property types such as apartments.  

- **Countryside:**  
  Displays a **high median price** with a **broad spread**, suggesting both low-cost and high-end properties exist in rural areas.  

---

### Summary
- **Location** plays a more critical role in determining property prices than **size**.  
- **Suburban and countryside** properties tend to command higher prices and show more variability.  
- **Downtown** areas have lower but more stable prices, reflecting greater market uniformity.

---


### Correlation Analysis

```python
# Proceeding with Correlation Analysis

# Calculate and visualize the correlation between Size_sqft and SalePrice
correlation = properties_df[['Size_sqft', 'SalePrice']].corr()

# Initialize the figure
plt.figure(figsize=(10, 5))

# Plotting the heatmap for correlation
sns.heatmap(correlation, annot=True, cmap='viridis', vmin=-1, vmax=1)
plt.title('Correlation between Property Size and Sale Price')

plt.show()
```

###  3. Correlation Between Property Size and Sale Price
A correlation analysis between **property size (sqft)** and **sale price** yielded a value of **-0.031**.

- This indicates a **very weak negative correlation**, effectively meaning **no meaningful linear relationship** between property size and price in the dataset.  
- In other words, **larger properties are not necessarily more expensive**, and smaller ones arent consistently cheaper.  
- This supports the earlier observation from the scatter plot that **other factors  particularly location  likely play a stronger role** in determining property value.

---

 **Insight:**  
While we might expect property size to correlate positively with sale price, this result suggests that market pricing in this dataset is **driven more by location, property type, and qualitative features** rather than sheer size.


```python
# Proceeding with Outlier Detection

# Initialize the figure
plt.figure(figsize=(15, 7))

# Box plots for Size_sqft and SalePrice to detect outliers
plt.subplot(1, 2, 1)
sns.boxplot(data=properties_df, y='Size_sqft', palette='viridis')
plt.title('Box Plot for Property Size (sqft)')
plt.ylabel('Size (sqft)')

plt.subplot(1, 2, 2)
sns.boxplot(data=properties_df, y='SalePrice', palette='viridis')
plt.title('Box Plot for Sale Price')
plt.ylabel('Sale Price')

plt.tight_layout()
plt.show()
```

```python
# Identifying potential outliers for SalePrice based on IQR
Q1 = properties_df['SalePrice'].quantile(0.25)
Q3 = properties_df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the potential outliers
outliers = properties_df[(properties_df['SalePrice'] < lower_bound) | (properties_df['SalePrice'] > upper_bound)]

# Plotting the scatter plot with potential outliers highlighted
plt.figure(figsize=(15, 7))

# Scatter plot for regular data points
for location, color in zip(['Suburb', 'Downtown', 'Countryside'], [colors[4], colors[7], colors[2]]):
    subset = properties_df[properties_df['Location'] == location]
    plt.scatter(subset['Size_sqft'], subset['SalePrice'], color=color, label=location, marker='^', alpha=0.7)

# Highlighting the outliers
plt.scatter(outliers['Size_sqft'], outliers['SalePrice'], color='red', label='Potential Outliers', s=150, edgecolors='black')

plt.title('Relationship between Property Size and Sale Price with Potential Outliers Highlighted')
plt.xlabel('Size (sqft)')
plt.ylabel('Sale Price')
plt.legend()

plt.show()
```
