```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('data/patients.csv')
raw_data.head()
```

```python
# Fill missing values in the Diagnosis column with "Unknown"
raw_data['Diagnosis'].fillna('Unknown', inplace=True)
```

```python
raw_data
```

```python
# drop duplicate
raw_data.drop_duplicates(inplace=True)
```

```python
raw_data
```

Turns out theres no duplicate :|

```python
# Ensure Age is a positive integer
raw_data['Age'] = raw_data['Age'].astype(int)
raw_data = raw_data[raw_data['Age'] > 0]

# Ensure LabResult is a positive integer
raw_data['LabResult'] = raw_data['LabResult'].astype(int)
raw_data = raw_data[raw_data['LabResult'] > 0]
```

```python
# get unique values tp encode in diagnosis
raw_data['Diagnosis'].unique()
```

```python
# Encode Diagnosis
raw_data['Diagnosis_Encoded'] = raw_data['Diagnosis'].map({'Unknown':0,'Diabetes':1,'Hypertension':2,'Flu':3,'Common Cold':4})
raw_data
```
